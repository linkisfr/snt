    <strong>Exercice 1</strong> <em>Fonction à un paramètre</em><br>   
La fréquence cardiaque maximale (FCM) est le nombre de battements par minute que le coeur atteint lors des plus grands efforts physiques. La FCM d'une personne dépend de son âge. Elle peut être estimée à l'aide de la formule $FCM = 192-0,007a^2$, où la variable $a$ est l'âge de la personne (en années).<br>
    Ecrire une fonction <code>FCM(age)</code> qui prend en paramètre l'âge d'une personne et renvoie sa fréquence cardiaque maximale.<br>
Effectuer ensuite un appel à cette fonction pour connaître votre fréquence cardiaque maximale.
</div>

```python
# Votre fonction ici : (attention à ne pas utiliser d'instruction print dans la définition de la fonction!)
def FCM(age):
    return 192 - 0.007 * age ** 2

# Votre test ici : (ici, vous pouvez utiliser print)
FCM(16)
```

 <div style="padding:10px ;border:solid 4px lightblue; border-radius: 10px; background-color:aliceblue;">
    <strong>Exercice 2</strong> <em>Fonction à deux paramètres</em><br>
 L'indice de Masse Corporelle (IMC) est une grandeur internationale qui permet de déterminer la corpulence d'une personne entre 18 ans et 65 ans. Il se calcule avec la formule : $IMC = \frac{\text{masse}}{\text{taille}^2}$, la masse étant exprimée en kilogrammes et la taille en mètres.<br>
    1) Ecrire une fonction <code>IMC(masse,taille)</code> qui renvoie l'IMC correspondant à la masse et à la taille données en argument.<br>
    2) La corpulence d'une personne est jugée normale lorsque son IMC vérifie : $18,5\leq IMC\lt25$.<br>Ecrire un programme qui demande à l'utilisateur de saisir sa masse et sa taille, et qui affiche si sa corpulence est normale ou non.
</div>

```python
# Votre fonction ici :
def IMC(masse, taille):
    return masse / (taille **2)
```

```python
# Test de la fonction
# on vérifie si l'écart avec la bonne valeur est inférieur à un millionième
assert(abs(IMC(70,1.8) - 21.604938271604937) < 1e-6) 
```

```python
# Votre programme qui fait appel à la fonction ici :
masse = float(input("Quelle est votre masse en kg ? "))
taille = float(input("Quelle est votre taille en mètres ? "))
imc = IMC(masse, taille)
if 18.5 <= imc and imc < 25:
    print("Corpulence normale")
elif imc < 18.5:
    print("IMC trop bas")
else:
    print("IMC trop élevé")
```

   Ecrire une fonction <code>somme</code> qui prend trois paramètres numériques et qui renvoie la somme des trois paramètres.<br>
    Ecrire ensuite une fonction <code>moyenne</code> qui prend aussi trois paramètres et qui renvoie la moyenne des trois paramètres. Cette fonction <code>moyenne</code> devra faire appel à la fonction <code>somme</code>.</div>
</div>

```python
# Votre fonction somme ici :
def somme(a, b, c):
    return a + b + c
```

```python
# Test
assert(somme(1,2,3) == 6)
```

```python
# Votre fonction moyenne ici :
def moyenne(a, b, c):
    return somme(a, b, c) / 3
```

```python
# Test
assert(moyenne(4,6,11) == 7.0)
```

<div style="padding:10px ;border:solid 4px lightblue; border-radius: 10px; background-color:aliceblue;"><strong>Exercice 6</strong> <em>L'erreur mystérieuse</em><br>
    Dans le programme ci-dessous, on a  créé une variable <code>compteur</code> dont le rôle est de compter combien de fois la fonction <code>maFonction</code> a été appelée. Le programme ci-dessous devrait normalement afficher 2, mais c'est un message d'erreur qui s'affiche.<br>
    Expliquer cette erreur et la corriger.
</div>

```python
compteur = 0

def maFonction():    
    compteur += 1
    return
    
print(maFonction())
print(maFonction())
print(compteur)
```

```python
# Par défaut, la valeur compteur est considérée comme une variable locale à la fonction
# Comme cette variable n'a pas été initialisée dans le corps de la fonction, on obtient une erreur.
# En réalité, on souhaite que compteur soit une variable globale, accessible partout, en dehors de la fonction
# et dans le corps de la fonction. Pour cela, on doit déclarer cette variable comme étant globale au début du corps de 
# la fonction, avec le mot-clé global

# Version corrigée : 
compteur = 0

def maFonction():
    global compteur   # on déclare ici la variable compteur comme étant globale
    compteur += 1
    return
    
maFonction()
maFonction()
print(compteur)
```

```python
# Votre fonction ligne ici :
def ligne(n, caractère):
    for i in range(n):
        print(caractère, end = "")
    print()  # on rajoute un saut de ligne
```
