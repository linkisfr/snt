   Le programme ci-dessous doit permettre de calculer le nombre de cartes du chateau de cartes.
    Il comporte plusieurs erreurs. <br>
    Corrigez-le pour qu'il fonctionne correctement.<br>
    <div style="transform:rotate(180deg);font-size:10px;text-align:right">Remarque : Le nombre de cartes du chateau est cent.</div>


```python
nbCartesObliquesParEtage = 2
nbCartesTotal = 0
for étage in range(1,8):    
    nbCartesTotal += nbCartesObliquesParEtage
    nbCartesTotal += étage
    nbCartesObliquesParEtage += 2
nbCartesTotal += nbCartesObliquesParEtage
print(nbCartesTotal)    
```
