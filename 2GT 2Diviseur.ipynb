## 6. Remainder <a name="reste"></a>
On considère la fonction `remainder(a, b)` où les arguments `a` et `b` sont deux entiers naturels.  
<b>a.</b> Que renvoie `remainder(33, 10)` ?  
<b>b.</b>  Que renvoie `remainder(40, 5)`  ?  
<b>c.</b>  De façon générale que fait `remainder(a, b)` ?

```python
def remainder(a, b):
    r = a
    while r >= b:
        r = r - b
    return r
```

```python
remainder(33, 10)
```

## 7. Smallest divisor <a name="small"></a>

<b>b.</b> De façon générale que fait `smallest_divisor(n)` ?

```python
def smallest_divisor(n):
    d = 2
    while d <= n:
        if n % d == 0:
            return d
        d = d + 1
```

## 8. List of divisors 
<b>a.</b> Que renvoie `list_divisors(33)` ?  Que renvoie `list_divisors(18)` ?  
<b>b.</b> De façon générale que renvoie `list_divisors(n)` ?                                    

```python
def list_divisors(n):
    L = []
    d = 2
    while d <= n:
        if n % d == 0:
            L.append(d)
        d = d + 1
    return L
```

```python
list_divisors(33)
```

<b>Remarque :</b> Si c'est trop facile pour vous voilà une version `list_divisors2(n)` qui vous intéressera!

```python
def list_divisors2(n):
    return [entier for entier in range(1, n+1) if n % entier == 0]
```

```python
list_divisors2(33)
```
