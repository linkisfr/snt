## 1. Comment définir une fonction? <a name="5"></a>


Si l'on veut transcrire en langage python la fonction affine définie par $f(x)=-2x+3$, il faut écrire ce qui est dans le scrpit suivant.  
Exécutez le script pour voir ce qui se passe.

```python
def f(x): # En python, lorsqu'une ligne de code se termine par :, la ligne suivante est décalée. On parle d' indentation
    return -2*x + 3
```

**Remarque :** inutile d'éxécuter plusieurs fois la cellule précédente, à part le numéro dans **Entrée[]**  (qui indique combien de fois vous avez exécuté la cellule) rien ne changera.

## 3. Boucle <a name="7"></a>
**a** Une fois la fonction définie, si on veut calculer les images de 0, 1, 2, ... jusqu'à 10, il suffit d'utiliser une boucle.  Exécutez la cellule ci-dessous pour voir le résultat.

```python
x = 0 # On initialise x c'est-a-dire on indique sa valeur de départ (ici j'ai choisi 0)
while x <= 10: # While signifie tant que
    print (f(x))
    x = x + 1  # On change la valeur de x, on augmente x de 1
```
