<h2 style="text-decoration:underline;" id="partie1">1- La boucle while</h2>

<p style="text-align:justify;">La boucle <code>for</code> a l'avantage d'être simple à utiliser, mais elle nécessite de connaître à l'avance (avant d'entrer dans la boucle) le nombre de tours de boucles qui vont être effectués. Dans certaines situations, on souhaiterait demander au programme de sortir de la boucle lorsqu'une certaine condition est réalisée, par exemple lorsqu'une valeur dépasse un certain seuil, ou lorsque l'utilisateur a saisi le bon mot de passe ou cliqué sur un bouton. La boucle <code>while</code> permet de répondre à ce besoin.</p>

**Syntaxe de la boucle while** <br><br>
<p style="border:solid;padding:5px">
   <code>while condition:
    bloc d'instructions A       
bloc d'instructions B
</code></p>

<p style="text-align:justify;">Tant que la condition est vraie, le bloc d'instructions A est exécuté, généralement plusieurs fois. Les instructions du bloc A agissent sur la condition pour la rendre fausse au bout d'un certain nombre de tours. Alors le programme sort de la boucle et se poursuit  avec l'exécution du bloc B.</p>

<strong>Remarques : </strong>
<ul style="text-align:justify;">
    <li>
La boucle <code>while</code> est parfois appelée "boucle non bornée" (nombre de tours non connu à l'avance, pouvant éventuellement dépasser n'importe quelle valeur) par opposition à la boucle <code>for</code> qui est appelée "boucle bornée" (le nombre de tours de boucle ne dépassera pas une valeur fixée à l'avance).
    </li>
    <li>
Il est possible que la boucle ne soit jamais exécutée, si la condition est initialement fausse.
    </li>
    <li>
       Il est possible que le programme boucle indéfiniment et ne s'arrête jamais, lorsque la condition est toujours vraie. On parle alors de <strong>boucle infinie</strong>. <br>
        Dans Capytale, il est recommandé de sauvegarder son travail régulièrement car il n'est pas toujours possible d'interrompre le programme en cas de boucle infinie.
    </li>
    <li>
        On peut se passer complètement des boucles <code>for</code> et les remplacer par des boucles <code>while</code> comme le montrent les deux programmes équivalents ci-dessous : 
    </li>
    
</ul>

```python
for i in range(5):
    print("Coucou")
```

```python
i = 0
while i < 5:
    print("Coucou")
    i = i + 1
```

<p style="text-align:justify;">Comme on le voit, l'utilisation de la boucle <code>while</code> nécessite d'écrire deux lignes de plus car il faut gérer soi-même la variable de boucle (son initialisation et son incrémentation).<br>
Lorsque cela est possible (c'est-à-dire si le nombre de tours est connu à l'avance), il est donc préférable d'utiliser une boucle <code>for</code> pour améliorer la lisibilité du code.</p>

<h4>Dépassement d'un seuil</h4>
<p style="text-align:justify;">La boucle <code>while</code> est souvent utilisée (notamment en mathématiques) pour déterminer le rang à partir duquel un certain seuil sera dépassé ou atteint.<br>
Voci un premier exemple :<br>
On dispose d'un capital de 1 000€, et chaque année, on perçoit 5% d'intérêts composés et on ajoute 500€ sur le compte.<br>
Au bout de combien d'années la valeur du compte dépassera-t-elle les 10 000€?</p>

```python
from tutor import tutor

def nombreAnnées(capital):
    année = 0
    while capital <= 10000:
        capital = capital * 1.05 + 500 
        année += 1
    return année

nombreAnnées(1000)

tutor()
```

<h4>Attente de la réalisation d'un événement</h4>
<p style="text-align:justify;">Plus généralement, on utilise la boucle <code>while</code> lorsqu'on souhaite attendre qu'une certain événement soit réalisé avant de continuer (saisie du bon mot de passe, déplacement de la souris, clic sur un bouton, détection d'un son dans le micro, etc...)<br>
Dans l'exemple ci-dessous, on attend que l'utilisateur saisisse la valeur 0 pour sortir de la boucle.</p>

```python
somme = 0
nombre = 1   # ici à la place de 1, on aurait pu mettre n'importe quel nombre (sauf 0)
while nombre != 0:
    nombre = int(input("Veuillez saisir le nombre de ventes réalisées (0 pour arrêter) : "))
    somme += nombre
print("La somme de vos ventes est",somme)    
```

<h2 style="text-decoration:underline;" id="partie2">2- L'instruction pass</h2>

Le code ci-dessous provoque une erreur car le corps de la boucle est vide :

```python
while 2 < 1:
```

<p style="text-align:justify;">Dans certains cas un peu particuliers, on peut avoir besoin de laisser temporairement vide le corps d'une boucle (for ou while) ou d'une fonction. Par exemple on peut souhaiter compléter plus tard le corps d'une fonction et le laisser vide sans que cela ne provoque une erreur. Pour cela, il existe en Python une instruction <code>pass</code> dont le rôle est ... de ne rien faire.<br>
Ainsi, le code ci-dessous ne provoque plus d'erreur :</p>

```python
while 2 < 1:
    pass
```

<h2 style="text-decoration:underline;" id="partie3">3- Sortie anticipée d'une boucle : l'instruction break</h2>

<p style="text-align:justify;">On a parfois besoin de sortir plus tôt que prévu d'une boucle en cas d'événement exceptionnel (par exemple si une erreur se produit). L'instruction <code>break</code> (qui doit être située dans le corps d'une boucle) permet de mettre fin immédiatement à la boucle. Le programme se poursuivra en exécutant les instructions situées juste après la boucle.</p>

```python
for i in range(5):
    âge = int(input("Veuillez saisir l'âge du passager n°"+str(i+1)+" : "))
    if âge < 12:
        print("Désolé, nous n'acceptons pas les enfants!")
        break  
```

<p style="text-align:justify;">Dans certaines situations, une boucle infinie combinée avec une instruction <code>break</code> peut permettre d'obtenir un programme un peu plus simple, comme dans le problème du mot de passe (exercice 4) qui est repris dans l'exercice suivant :</p>

<h2 style="text-decoration:underline;" id="partie4">4- Retour anticipé au début de la boucle : l'instruction continue</h2>

Dans une boucle (for ou while), l'instruction <code>continue</code> demande au programme de revenir immédiatement au début de la boucle pour effectuer le tour suivant, sans terminer le tour de boucle qui était en cours d'exécution.<br>
Cette instruction permet dans certains cas d'améliorer un peu la lisibilité du programme.

<div class="alert alert-info"><strong>Exercice 12 : </strong> Un hôtel comporte 19 chambres numérotées de 1 à 20, mais la chambre n°13 n'existe pas.<br>
Le programme ci-dessous affiche la liste de tous les numéros de chambres de l'hôtel.<br>
    Le petit défaut (qui concerne seulement la lisibilité) de ce programme est que l'instruction <code>print(i)</code> est reléguée au 2ème niveau d'indentation alors qu'il s'agit de l'instruction qui est effectuée dans les cas ordinaires (et non dans le cas exceptionnel de la chambre n°13).<br>
    Réécrire une version plus lisible de ce programme, dans laquelle  l'instruction <code>print(i)</code> sera écrite au 1er niveau d'indentation.
</div>

```python
for i in range(1,21):
    if i != 13:
        print(i)
```
