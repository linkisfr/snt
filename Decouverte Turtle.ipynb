## Découverte du module turtle

Le module turtle va nous permettre de dessiner des figures de façon très simple. Pour utiliser ce module, il faut :
* commencer le programme par l’instruction : `import turtle`.
* dessiner avec les commandes de la tortue (voir ci-dessous).
* terminer le programme par l’instruction : `turtle.done()`.

Ce module peut être vu comme une petite tortue qui porte une plume, soit la plume est abaissée ce qui est le cas initialement et dès que la tortue bouge, elle fait le tracé de ces déplacements, soit la plume est surélevée, à ce moment-là quand la tortue bouge, il n’y a pas de tracé qui est effectué.

Voici une liste de commandes turtle, les plus communes dont nous allons nous servir :

|Commande |Effet|
|---    |---    |
|turtle. reset() |On efface tout et on recommence|
|turtle.goto(x, y) |Aller à l’endroit de coordonnées x, y|
|turtle.forward(distance) |Avancer d’une distance donnée|
|turtle.backward(distance) |Reculer
|turtle.up() |Relever le crayon (pour pouvoir avancer sans dessiner)|
|turtle.down() |Abaisser le crayon (pour recommencer à dessiner)|
|turtle.color(couleur) |Couleur peut être ‘red’ , ‘blue’, etc.|
|turtle.left(angle) |Tourner à gauche d’un angle donné (exprimé en degrés)|
|turtle.right(angle) |Tourner à droite|
|turtle.width(épaisseur) |Choisir l’épaisseur du tracé|
|turtle.begin_fill() |Début de la zone fermée à colorier|
|turtle.end_fill() |Fin de la zone fermée à colorier|
|turtle.write(texte)|Texte doit être une chaîne de caractères|
|turtle.done() ou turtle.mainloop()| Attend que l’utilisateur ferme la fenêtre|
|turtle.hideturtle() |Cache la tortue|
|turtle.circle() |Rayon, angle (+ dans le sens trigo, - dans le sens horaire)|
|    |   |
