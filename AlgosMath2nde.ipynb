**10.** <a name="10"></a> Pour des données réelles ou issues d’une simulation, lire et comprendre une fonction écrite en Python renvoyant la moyenne $m$, l’écart type $s$, et la proportion d’éléments appartenant à $[m-2s;m+2s]$.

```python
from math import sqrt  # On a besoin de la racine carrée pour l'écart type

def moyenne(liste):
    nombre_d_elements = 0  # Au commencement, on n’a traité aucun élément de la liste.
    somme = 0              # Au commencement, la somme des éléments vaut 0.
    for element in liste:  # Pour chaque élément de la liste,
        somme = somme + element  # On ajoute à la somme des éléments le nouvel élément considéré.
        nombre_d_elements = nombre_d_elements + 1  # On indique qu’on a traité un élément supplémentaire.
    return somme/nombre_d_elements  # Moyenne : on divise la somme totale par le nombre d’éléments.

def ecart_type(liste):
    nombre_d_elements = 0  # Au commencement, on n’a traité aucun élément de la liste.
    somme = 0              # Au commencement, la somme à calculer vaut 0.
    x = moyenne(liste)     # Notons x la moyenne des éléments de la liste.
    for element in liste:  # Pour chaque élément de la liste,
        somme = somme + (element - x)**2  # On ajoute à la somme la partie de la formule de l’élément considéré.
        nombre_d_elements = nombre_d_elements + 1  # On indique qu’on a traité un élément supplémentaire.
    variance = somme/nombre_d_elements  # Variance : cette somme divisée par le nombre d’éléments.
    return sqrt(variance)  # Écart type = racine carrée de la variance.

def proportion(liste):
    nombre_d_elements = 0  # Au commencement, on n’a traité aucun élément de la liste.
    n = 0                  # Au commencement, il n’y a (donc) aucun élément appartenant à l’intervalle
    x = moyenne(liste)     # Notons x la moyenne des éléments de la liste
    s = ecart_type(liste)  # Notons s l’écart type des éléments de la liste
    for element in liste:  # Pour chaque élément de la liste,
        nombre_d_elements = nombre_d_elements + 1  # On indique qu’on a traité un élément supplémentaire,
        if x - 2*s <= element <= x + 2*s:  # et si cet élément est compris dans l’intervalle,
            n = n + 1  # alors on ajoute 1 au nombre d’éléments appartenant à l’intervalle.
    return n/nombre_d_elements  # Proportion = nombre d’éléments dans l’intervalle ÷ nombre total d’éléments
```

**11.** <a name="11"></a> Lire et comprendre une fonction Python renvoyant le nombre ou la fréquence de succès dans un échantillon de taille $n$ pour une expérience aléatoire à deux issues. 

```python
from random import random()

def frequence_de_succes(nombre_de_lancers, probabilite_succes):
    n = 0  # Au commencement, on n’a aucun succès
    for _ in range(nombre_de_lancers):  # On répète un certain nombre de lancers,
        if random() < probabilite_succes:  # et si on obtient un succès,
            n = n + 1  # on indique qu’on a un succès complémentaire.
    return n/nombre_de_lancers  # Fréquence = nombre de succès ÷ nombre de lancers
```

```python
frequence_de_succes(5, 1/6)
```

**12.** <a name="12"></a>Observer la loi des grands nombres à l’aide d’une simulation sur Python ou tableur.  

En statistiques, la loi des grands nombres exprime le fait que les caractéristiques d’un échantillon aléatoire se rapprochent des caractéristiques statistiques de la population lorsque la taille de l’échantillon augmente. Autrement dit, la fréquence (dans la pratique) de succès d’un évènement sur l’échantillon s’approche de la probabilité (théorique) de cet évènement sur la population.

```python
for i in [10, 100, 1000, 10000, 100000]:  # Pour cinq échantillons de tailles de plus en plus grandes,
    print(frequence_de_succes(i, 1/4))    # on indique la fréquence de succès d’une expérience où P(succès) = 1/4.
```
